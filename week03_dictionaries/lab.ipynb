### Concept 2.1: Advanced Slicing

Slicing syntax: `sequence[start:stop:step]`

- Negative indices count from the end
- Negative step reverses direction
- Omitted values have defaults: start=0, stop=len, step=1

```python
# Security log entries (timestamps)
log_times = ["09:15", "09:22", "09:45", "10:01", "10:15", "10:30", "11:00"]

# Get last 3 entries (most recent)
print("Last 3:", log_times[-3:])

# Reverse the entire list
print("Reversed:", log_times[::-1])

# Every other entry starting from second
print("Every other (from index 1):", log_times[1::2])
```

### Concept 2.2: Tuple Unpacking and the * Operator

Unpacking allows you to assign multiple variables at once. The `*` operator captures "the rest."

```python
# Network connection tuple: (source_ip, source_port, dest_ip, dest_port, protocol)
connection = ("192.168.1.100", 54321, "10.0.0.1", 443, "TCP")

# Unpack all values
src_ip, src_port, dst_ip, dst_port, proto = connection
print("Destination: {}:{}".format(dst_ip, dst_port))

# Capture middle elements with *
first, *middle, last = connection
print("First:", first)
print("Middle:", middle)
print("Last:", last)
```

### Concept 2.3: Dictionary - get(), setdefault(), and update()

These methods prevent KeyError exceptions and enable cleaner code.

```python
# Failed login attempts by IP
failed_logins = {
    "192.168.1.50": 3,
    "10.0.0.25": 7
}

# WRONG WAY - causes KeyError if IP not found
# count = failed_logins["172.16.0.1"]  # KeyError!

# RIGHT WAY - get() with default value
count = failed_logins.get("172.16.0.1", 0)
print("Unknown IP attempts:", count)

# setdefault() - get value OR set it if missing
failed_logins.setdefault("172.16.0.1", 0)
print("After setdefault:", failed_logins)

# update() - merge dictionaries
new_data = {"192.168.1.50": 5, "8.8.8.8": 1}
failed_logins.update(new_data)
print("After update:", failed_logins)
```

### Concept 2.4: Set Operations for Security Analysis

Sets are perfect for comparing collections: finding common elements, differences, or unique items.

```python
# IPs that accessed the server yesterday vs today
yesterday_ips = {"192.168.1.10", "192.168.1.20", "10.0.0.5", "172.16.0.1"}
today_ips = {"192.168.1.10", "192.168.1.30", "10.0.0.5", "8.8.8.8"}

# Intersection: IPs that accessed BOTH days (regular users?)
print("Both days:", yesterday_ips & today_ips)

# Difference: NEW IPs today (potential threats?)
print("New today:", today_ips - yesterday_ips)

# Symmetric difference: IPs that accessed ONLY one day
print("Only one day:", yesterday_ips ^ today_ips)

# Union: ALL unique IPs across both days
print("All IPs:", yesterday_ips | today_ips)
```

### Concept 2.5: Tricky Behavior - Default Mutable Arguments

**WARNING:** This is a common Python pitfall that can cause security bugs!

```python
# DANGEROUS: Mutable default argument
def add_user_BAD(username, permissions=[]):
    permissions.append("read")
    return {"user": username, "perms": permissions}

user1 = add_user_BAD("alice")
user2 = add_user_BAD("bob")

print("User 1 permissions:", user1["perms"])
print("User 2 permissions:", user2["perms"])
# Notice anything wrong?

# Are they the same object?
print("Same object?", user1["perms"] is user2["perms"])
```
